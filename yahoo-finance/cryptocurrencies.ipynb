# CRYPTOCURRENCIES

```python
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
```

```python
ticker = ["BTC-USD", "ETH-USD"]
```

```python
crypt = yf.download(ticker, period='5y').Close
crypt
```

```python
norm = crypt.div(crypt.iloc[0]).mul(100)
```

```python
norm.plot()
plt.grid()
plt.show()
```
