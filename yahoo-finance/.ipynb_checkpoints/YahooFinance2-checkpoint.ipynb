## setting specific time periods

```python
df = yf.download(ticker, start='2019-01-01', end='2023-06-01')
```

```python
# intraday cannot extend last 60 days
df.head()
```

```python
df = yf.download(ticker, period='ytd')
df.head()
```

```python
df = yf.download(ticker, period='1mo')
# Valid periods: 1d,5d,1mo,3mo,6mo,1y,2y,5y,10y,ytd,max
df.head()
```

## Frequency settings (intraday)

```python
df = yf.download(ticker, period='3mo', interval='1h')
```

```python
# Valid intervals: 1m,2m,5m,15m,30m,60m,90m,1h,1d,5d,1wk,1mo,3mo
```

```python
df.head(10)
# for intervals more than 1 day, download for 1 day and convert internally
```

```python
df.describe()
```

```python
df = yf.download(ticker, period='3mo', interval='1h', prepost=True)
```

```python
df.head(20)
```

## stock splits and dividends

```python
ticker = 'AAPL'
```

```python
aapl = yf.download(ticker, period='10y', actions=True)
```

```python
aapl[aapl['Dividends'] != 0.0].tail(10)
```

```python
aapl.loc['2023-08-08':'2023-08-20']
```

```python
aapl.loc['2023-08-08':'2023-08-20'].diff()
```

```python
aapl[aapl['Stock Splits'] != 0.0]
```

```python
aapl.loc['2020-08-25':'2020-09-05']
```

## exporting files to csv/excel

```python
ticker = 'GE'
```

```python
GE = yf.download(ticker, period='5y')
```

```python
GE.head()
```

```python
GE.to_csv("GE.csv")
```

```python
df = pd.read_csv("GE.csv", parse_dates=['Date'], index_col="Date")
df.head()
```

```python
df.to_excel("GE.xls", engine='openpyxl')
```

```python
pd.read_excel("GE.xls", parse_dates=['Date'], index_col="Date")
```
