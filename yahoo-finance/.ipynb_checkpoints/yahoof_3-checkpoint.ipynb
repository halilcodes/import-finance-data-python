```python
import yfinance as yf
import pandas as pd
```

```python
tickers = ["GE", "AAPL", "META", "TSLA"]
```

```python
yf.download(tickers, period='5y')
```

```python
yf.download(tickers, period='5y', group_by='ticker')
```

```python
stocks = yf.download(tickers, period='5y').Close
```

```python
stocks.head()
```

```python
import matplotlib.pyplot as plt
```

```python
stocks.plot()
plt.show()
```
