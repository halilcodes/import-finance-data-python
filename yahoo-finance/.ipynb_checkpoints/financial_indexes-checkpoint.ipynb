# Importing Financial Indexes

```python
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
```

```python
tickers = ["^DJI", "^GSPC"]
```

```python
yf.download(tickers, period='5y')
```

```python
indexes = yf.download(tickers, period='5y').Close
```

```python
indexes.head()
```

```python
norm = indexes.div(indexes.iloc[0]).mul(100)
norm
```

```python
norm.plot()
plt.grid()
plt.show()
```
